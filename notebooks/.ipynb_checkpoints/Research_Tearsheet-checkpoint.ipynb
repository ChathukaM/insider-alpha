# Insider Cluster-Buy Signal: Does It Still Work?

**Research Question:** When multiple corporate insiders purchase their company's stock within a short window, does it predict abnormal returns?

**Data:** SEC EDGAR Form 4 filings, 2010–2024 (729,122 raw transactions)

**Methodology:** Filter for high-conviction open-market purchases by officers and directors, identify cluster buys (3+ distinct insiders within 30 days, $500k+ combined value), backtest with 60-day holding period, benchmark against S&P 500

**Finding:** The signal produces a 60.7% hit rate with a statistically significant abnormal return of 2.37% (t-stat: 2.67) across the full sample. However, walk-forward validation reveals that the outperformance is concentrated in crisis periods — particularly the 2020 COVID crash. Excluding 2020, abnormal returns are near zero. The insider buying anomaly appears to function as a crisis indicator rather than an all-weather alpha source.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 12

trades = pd.read_csv('../data/processed/trades.csv')
trades['signal_date'] = pd.to_datetime(trades['signal_date'])
trades['entry_date'] = pd.to_datetime(trades['entry_date'])
trades['exit_date'] = pd.to_datetime(trades['exit_date'])
trades['year'] = trades['signal_date'].dt.year

signals = pd.read_csv('../data/processed/signals.csv')
purchases = pd.read_csv('../data/processed/insider_purchases.csv', low_memory=False)

print(f'Total trades executed: {len(trades)}')
print(f'Date range: {trades["signal_date"].min().date()} to {trades["signal_date"].max().date()}')
```

## 1. Data Pipeline Overview

The pipeline processes SEC Form 4 filings through progressive filtering to isolate high-conviction insider purchases.

How are cluster buy signals distributed over time? We expect more signals during market downturns when insiders buy the dip.

```python
# signals per year
yearly_signals = trades.groupby('year').size()

fig, ax = plt.subplots(figsize=(10, 5))
colors = ['#E74C3C' if y == 2020 else '#2C3E50' for y in yearly_signals.index]
ax.bar(yearly_signals.index, yearly_signals.values, color=colors)
ax.set_xlabel('Year')
ax.set_ylabel('Number of Cluster Buy Signals')
ax.set_title('Trading Signals by Year')
ax.set_xticks(yearly_signals.index)

# annotate 2020
ax.annotate('COVID crash\n(peak insider buying)',
            xy=(2020, yearly_signals.loc[2020]),
            xytext=(2020, yearly_signals.loc[2020] + 20),
            ha='center', fontsize=9, color='#E74C3C',
            arrowprops=dict(arrowstyle='->', color='#E74C3C'))

plt.tight_layout()
plt.show()
```

## 3. Overall Backtest Results

```python
# summary statistics
total = len(trades)
winners = (trades['stock_return'] > 0).sum()
losers = total - winners

summary = {
    'Total Trades': total,
    'Winners': winners,
    'Losers': losers,
    'Hit Rate': f"{winners/total:.1%}",
    'Avg Return': f"{trades['stock_return'].mean():.2%}",
    'Avg Abnormal Return': f"{trades['abnormal_return'].mean():.2%}",
    'Median Return': f"{trades['stock_return'].median():.2%}",
    'Avg Winner': f"{trades[trades['stock_return']>0]['stock_return'].mean():.2%}",
    'Avg Loser': f"{trades[trades['stock_return']<=0]['stock_return'].mean():.2%}",
    'Sharpe Ratio': f"{(trades['stock_return'].mean() / trades['stock_return'].std()) * np.sqrt(252/60):.2f}",
    'Abnormal Return t-stat': f"{trades['abnormal_return'].mean() / (trades['abnormal_return'].std() / np.sqrt(total)):.2f}"
}

summary_df = pd.DataFrame(list(summary.items()), columns=['Metric', 'Value'])
print(summary_df.to_string(index=False))
```

## 4. Yearly Performance Breakdown

```python
yearly = trades.groupby('year').agg(
    trades=('stock_return', 'count'),
    hit_rate=('stock_return', lambda x: (x > 0).mean()),
    avg_return=('stock_return', 'mean'),
    avg_abnormal=('abnormal_return', 'mean')
).round(4)

yearly['hit_rate'] = yearly['hit_rate'].apply(lambda x: f"{x:.1%}")
yearly['avg_return'] = yearly['avg_return'].apply(lambda x: f"{x:.2%}")
yearly['avg_abnormal'] = yearly['avg_abnormal'].apply(lambda x: f"{x:.2%}")

print(yearly.to_string())
```

```python
# yearly abnormal returns chart
yearly_raw = trades.groupby('year').agg(
    avg_abnormal=('abnormal_return', 'mean')
)

fig, ax = plt.subplots(figsize=(10, 5))
colors = ['#27AE60' if x > 0 else '#E74C3C' for x in yearly_raw['avg_abnormal']]
ax.bar(yearly_raw.index, yearly_raw['avg_abnormal'] * 100, color=colors)
ax.axhline(y=0, color='black', linewidth=0.8)
ax.set_xlabel('Year')
ax.set_ylabel('Average Abnormal Return (%)')
ax.set_title('Abnormal Returns by Year (60-Day Hold)')
ax.set_xticks(yearly_raw.index)

plt.tight_layout()
plt.show()
```

## 5. Holding Period Sensitivity

The choice of holding period significantly affects results. We test 20, 40, 60, 90, and 120 trading days to identify the optimal horizon.

```python
# holding period comparison
# these values were computed during parameter exploration
holding_periods = {
    'Period': ['20 days', '40 days', '60 days', '90 days', '120 days'],
    'Avg Return': ['0.75%', '2.84%', '6.94%', '8.63%', '10.64%'],
    'Avg Abnormal': ['-1.08%', '-0.33%', '2.37%', '1.71%', '1.22%'],
    'Hit Rate': ['52.9%', '53.9%', '60.7%', '55.9%', '57.8%']
}

hp_df = pd.DataFrame(holding_periods)
print('Holding Period Sensitivity Analysis:')
print(hp_df.to_string(index=False))
print('\nThe 60-day holding period maximises abnormal returns and hit rate.')
print('This is consistent with insiders having a quarterly outlook on their companies.')
```

## 6. Walk-Forward Validation

To test for overfitting, we split the data into an in-sample discovery period (2010–2017) and an out-of-sample test period (2018–2024). Parameters were selected on the full dataset, so this is a conservative check — a true out-of-sample test would require real-time deployment.

```python
def compute_stats(df, label):
    total = len(df)
    hr = (df['stock_return'] > 0).mean()
    avg_ret = df['stock_return'].mean()
    avg_abn = df['abnormal_return'].mean()
    t_stat = avg_abn / (df['abnormal_return'].std() / np.sqrt(total))
    return {
        'Period': label,
        'Trades': total,
        'Hit Rate': f'{hr:.1%}',
        'Avg Return': f'{avg_ret:.2%}',
        'Avg Abnormal': f'{avg_abn:.2%}',
        't-stat': f'{t_stat:.2f}'
    }

in_sample = trades[trades['year'] <= 2017]
out_sample = trades[trades['year'] >= 2018]
out_no_2020 = out_sample[out_sample['year'] != 2020]
full_no_2020 = trades[trades['year'] != 2020]

wf_results = [
    compute_stats(trades, 'Full Sample (2010-2024)'),
    compute_stats(in_sample, 'In-Sample (2010-2017)'),
    compute_stats(out_sample, 'Out-of-Sample (2018-2024)'),
    compute_stats(out_no_2020, 'Out-of-Sample ex-2020'),
    compute_stats(full_no_2020, 'Full Sample ex-2020'),
]

wf_df = pd.DataFrame(wf_results)
print('Walk-Forward Validation Results:')
print(wf_df.to_string(index=False))
```

## 9. Conclusion

### Key Findings

1. **The insider cluster-buy signal has a consistent hit rate above 50%** across all sample periods, confirming that corporate insiders do identify undervalued stocks better than random selection.

2. **Abnormal returns are concentrated in crisis periods.** The 2020 COVID crash produced 17.2% average abnormal returns with a 77% hit rate. Excluding this period, abnormal returns are near zero.

3. **The 60-day holding period is optimal.** At 20 days, the signal shows negative abnormal returns — insiders think in quarters, not weeks. The signal peaks at 60 days and decays thereafter.

4. **The anomaly has been largely arbitraged away in normal markets.** Walk-forward validation shows the signal does not generate consistent alpha outside of extreme market dislocations.

### Interpretation

The economic logic is sound: during market panics, prices disconnect from fundamentals and insiders — who understand their companies better than anyone — identify genuine mispricings. In calm markets, prices are more efficient and the informational advantage of insiders is smaller, consistent with the efficient market hypothesis.

### Limitations

- **Survivorship bias**: 406 out of 928 tickers failed to download from Yahoo Finance, likely delisted or acquired companies. The surviving tickers may have a positive bias.
- **Parameter selection**: The 60-day holding period and filter thresholds were selected on the full dataset rather than a true hold-out sample.
- **Transaction costs**: The backtest assumes zero transaction costs and no market impact from trade execution.
- **Benchmark choice**: Using the S&P 500 as a benchmark may not perfectly match the risk profile of the traded stocks, many of which are small and mid-cap.

### Potential Extensions

- Combine insider signal with a market regime indicator (e.g., VIX level) to only trade during elevated volatility
- Add Fama-French factor regression to control for size, value, and momentum exposures
- Use CRSP/Compustat data to eliminate survivorship bias
- Test on international markets where the anomaly may be less arbitraged
